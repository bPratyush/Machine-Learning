```python
import numpy as np
import os
import torch
import scipy.io
from PIL import Image
import torch.nn.functional as F
from torch.utils.data import DataLoader

def preprocess_data(images_npy_file, ages_npy_file):
    images = np.load(images_npy_file)
    #Data Normalization
    images = images / 255.0 
    ages = np.load(ages_npy_file)
    return images, ages
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
images, ages = preprocess_data('/Users/bpratyush/Downloads/Machine Learning/CBAM/imdb-wiki-image-npy/imdb-wiki-image-data.npy', '/Users/bpratyush/Downloads/Machine Learning/CBAM/imdb-wiki-image-npy/age.npy')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(images, ages, test_size=0.2, random_state=42)
```

```python
from torch.utils.data import Dataset

class IMDBWIKIDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
train_dataset = IMDBWIKIDataset(X_train, y_train)
test_dataset = IMDBWIKIDataset(X_test, y_test)

train_dataloader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=32, shuffle=False)
```

```python
import matplotlib.pyplot as plt
plt.hist(ages, bins=20)
plt.title('Age distribution in dataset')
plt.xlabel('Age')
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class ChannelAttention(nn.Module):
    def __init__(self, in_planes, ratio=16):
        super(ChannelAttention, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)

        self.fc1 = nn.Conv2d(in_planes, in_planes // 16, 1, bias=False)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Conv2d(in_planes // 16, in_planes, 1, bias=False)

        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = self.fc2(self.relu1(self.fc1(self.avg_pool(x))))
        max_out = self.fc2(self.relu1(self.fc1(self.max_pool(x))))
        out = avg_out + max_out
        return self.sigmoid(out)

class SpatialAttention(nn.Module):
    def __init__(self, kernel_size=7):
        super(SpatialAttention, self).__init__()

        assert kernel_size in (3, 7), 'kernel size must be 3 or 7'
        padding = 3 if kernel_size == 7 else 1

        self.conv1 = nn.Conv2d(2, 1, kernel_size, padding=padding, bias=False)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        x = torch.cat([avg_out, max_out], dim=1)
        x = self.conv1(x)
        return self.sigmoid(x)

class CBAM(nn.Module):
    def __init__(self, in_planes):
        super(CBAM, self).__init__()
        self.ca = ChannelAttention(in_planes)
        self.sa = SpatialAttention()

    def forward(self, x):
        x = self.ca(x) * x
        x = self.sa(x) * x
        return x

class FaceAgeCNN(nn.Module):
    def __init__(self):
        super(FaceAgeCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1)
        self.cbam1 = CBAM(64)
        self.fc = nn.Linear(64*100*100, 1)

    def forward(self, x):
        out = F.relu(self.conv1(x))
        out = self.cbam1(out)
        out = out.view(out.size(0), -1)  
        out = self.fc(out)
        return out
```

```python
from torchviz import make_dot

model = FaceAgeCNN()
x = torch.randn(1, 1, 100, 100) 
y = model(x)
make_dot(y, params=dict(model.named_parameters())).render("FaceAgeCNN", format="png")

from IPython.display import Image
Image('FaceAgeCNN.png')
```

```python
model = FaceAgeCNN().to(device)

# Loss function
criterion = nn.MSELoss()
n_total_steps = len(train_dataloader)
# Reduced learning rate
learning_rate = 0.001 
num_epochs = 10
# Optimizer
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)

for epoch in range(num_epochs):
    for i, (images, ages) in enumerate(train_dataloader):
        images = images.view(-1, 1, 100, 100).float().to(device) 
        ages = ages.float().to(device)

        # Check for nan values in data
        assert torch.isnan(images).any() == False
        assert torch.isnan(ages).any() == False

        # Forward pass
        outputs = model(images)
        outputs = outputs.squeeze()  
        loss = criterion(outputs, ages)

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if (i+1) % 100 == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{n_total_steps}], Loss: {loss.item()}')
```

```python
model.eval()  # set the model to evaluation mode
with torch.no_grad():
    total_error = 0
    total_count = 0
    for images, ages in test_dataloader:
        images = images.view(-1, 1, 100, 100).float().to(device)  # Reshape images and convert to float
        ages = ages.float().to(device)
        outputs = model(images)
        outputs = outputs.squeeze()  # Remove the singleton dimension
        error = torch.abs(outputs - ages).sum().item()  # Calculate absolute error
        total_error += error
        total_count += ages.size(0)

    mae = total_error / total_count
    print('Mean Absolute Error on the test images: {}'.format(mae))
```

```python
import matplotlib.pyplot as plt

errors = []
for images, ages in test_dataloader:
    images = images.view(-1, 1, 100, 100).float().to(device)
    ages = ages.float().to(device)
    outputs = model(images)
    outputs = outputs.squeeze()
    error = torch.abs(outputs - ages)
    errors.extend(error.tolist())

plt.hist(errors, bins=20)
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.title('Distribution of Errors')
plt.show()
```
